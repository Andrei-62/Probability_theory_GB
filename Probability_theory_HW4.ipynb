## Задача4

Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
а). больше 182 см
б). больше 190 см
в). от 166 см до 190 см
г). от 166 см до 182 см
д). от 158 см до 190 см
е). не выше 150 см или не ниже 190 см
ё). не выше 150 см или не ниже 198 см
ж). ниже 166 см.

```python
def z_values(height):
    return (height-174)/8
```

```python
Z1 = z_values(182)
Z1
```

Так как Z = 1, значит рост смещается на 1 сигму от среднего роста

Исходя из правила 3 сигм

```python
print(f'Исходя из правила 3 сигм вероятность того, что случайным образом выбранный взрослый человек имеет рост больше 182 см = {(99.72-68)/2}%')
```

```python
print(z_values(190))
```

```python
print(f'Исходя из правила 3 сигм вероятность того, что случайным образом выбранный взрослый человек имеет рост больше 190 см = {(99.72-95.4)/2}%')
```

```python
print(z_values(166))
```

```python
print(f'Исходя из правила 3 сигм вероятность того, что случайным образом выбранный взрослый человек имеет рост от 166 см до 190 см = {68+((95.4-68)/2)}%')
```

```python
print(f'Исходя из правила 3 сигм вероятность того, что случайным образом выбранный взрослый человек имеет рост от 166 см до 182 см = {68}%')
```

```python
print(f'Исходя из правила 3 сигм вероятность того, что случайным образом выбранный взрослый человек имеет рост от 158 см до 190 см = {95.4}%')
```
